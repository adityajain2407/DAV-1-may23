```python
#task: Categorize a movie as risky or non-risky
#a movie is risky, if its budget is greater than the average revenue of all the movies of that director
```

```python
#groupby apply()
```

```python
def check_risky(df):

  risky_flag = df["budget"] >= df["revenue"].mean()
  df["risky"] = risky_flag
  return df
```

```python
data_risky = data.groupby("director_name").apply(check_risky)
```

```python
data_risky.head()
```

```python
data_risky[data_risky["risky"] == True].shape
```

```python
#Who is the most productive director?
```

```python
data.groupby("director_name")["year"].aggregate(["min", "max"])
```

```python
data.groupby("director_name")[["year", "title"]].aggregate(
    {
        "year": ["min", "max"],
        "title": "count"

    }
)
```

```python
#way2


data_agg = data.groupby("director_name")[["year", "title"]].aggregate(
    first_year = ("year", "min"),
    last_year = ("year", "max"),
    movie_count = ("title", "count")
)
```

```python
data_agg
```

```python
data_agg.reset_index()
```

```python
data_agg["years_active"] = data_agg["last_year"] - data_agg["first_year"] + 1
```

```python
data_agg.head()
```

```python
data_agg["movies_per_year"] = data_agg["movie_count"]/data_agg["years_active"]
```

```python
data_agg.head()
```

```python
#Display the data of the top 5 most productive directors
```

```python
data_agg.sort_values("movies_per_year", ascending = False)
```
