```python
def temp_avg(df):

  df["Temp_avg"] = df["Temperature"].mean()
  return df
```

```python
data_tidy = data_tidy.groupby("Drug_Name").apply(temp_avg)
```

```python
data_tidy.head()
```

```python
data_tidy[data_tidy["Drug_Name"] == "ketamine hydrochloride"].head()
```

```python
data_tidy["Temperature"].fillna(data_tidy["Temp_avg"], inplace = True)
```

```python
temp_points = [5, 20, 35, 50, 60]
temp_labels = ["low", "medium", "high", "very_high"]
```

```python
#cut
data_tidy["temp_bucket"] = pd.cut(data_tidy["Temperature"],
                                  bins = temp_points,
                                  labels = temp_labels
                                  )
```
