```python
data.head()
```

```python
def encode_gender(g):
  if g == "Male":
    return 0
  else:
    return 1
```

```python
data["gender"] = data["gender"].apply(encode_gender)
```

```python
def calc_profit(df):
  profit = df["revenue"] - df["budget"]
  return profit
```

```python
data["profit"] = data[["revenue", "budget"]].apply(calc_profit, axis = 1)
```
